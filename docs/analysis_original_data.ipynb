# Анализ исходных данных

```python
import pandas as pd
from rdkit import Chem
from IPython.display import display, Markdown
# ------------------------------------------------------
import sys
import os
current_dir = os.getcwd()
parent_dir = os.path.dirname(current_dir)
sys.path.append(parent_dir)
from utils.utils import start
start()

SDF_FILE_PATH = "../data/output_500.sdf"
```

```python
all_molecules_properties = []
suppl = Chem.SDMolSupplier(SDF_FILE_PATH)
for mol in suppl:
    all_molecules_properties.append(mol.GetPropsAsDict())
df = pd.DataFrame(all_molecules_properties)
display(df.head(3))
```

```python
print(f"Всего молекул в исходном наборе: {len(df)}")
```

```python
# Присуствует строка NaN во всех свойствах, удалим ее
df = df.dropna().reset_index(drop=True)
```

```python
# и еще раз проверим пропуски
print((df.isna().sum() / len(df) * 100).sort_values(ascending=False)[lambda x: x >0])
```

```python
suppl = Chem.SDMolSupplier(SDF_FILE_PATH)
mols = [mol for mol in suppl if mol is not None]
print(f"Всего молекул: {len(mols)}")
```

```python
display(mols[5])
```

```python
def show_props(molecule, index=None):    
    header_text = f"Анализ Молекулы №{index}"
    display(Markdown(f"### {header_text}"))
    display(molecule)
    props = molecule.GetPropsAsDict()
    props_df = pd.Series(props).to_frame(name='Значение')
    props_df.index.name = "Свойство"
    display(props_df)
    display(Markdown("---"))
show_props(mols[5], index=3)
```
